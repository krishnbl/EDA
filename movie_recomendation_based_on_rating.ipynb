```python
from sklearn.metrics.pairwise import cosine_similarity

# Create a user-item matrix
user_item_matrix = df.pivot_table(index='userId', columns='movie', values='rating')
```

```python
user_item_matrix
```

```python
# Fill missing values with 0
user_item_matrix_filled = user_item_matrix.fillna(0)
```

```python
user_item_matrix_filled
```

```python
# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix_filled)
```

```python
user_similarity
```

```python
# Convert to DataFrame for better readability
user_similarity_df = pd.DataFrame(user_similarity, index=df['userId'].unique(), columns=df['userId'].unique())
```

```python
def recommend_movies(movie_name,n_recommendations=5):
    """
    Function to recommend movies based on a given movie name using item-item similarity.
    
    Parameters:
    movie_name (str): The name of the movie for which recommendations are to be found.
    user_item_matrix (DataFrame): The user-item matrix with users as rows and movies as columns.
    n_recommendations (int): The number of movie recommendations to return. Default is 5.
    
    Returns:
    recommendations (DataFrame): A DataFrame with recommended movies and their similarity scores.
    """
    # Calculate the cosine similarity matrix for movies
    movie_similarity = cosine_similarity(user_item_matrix.T.fillna(0))
    movie_similarity_df = pd.DataFrame(movie_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)
    
    # Get the similarity scores for the specified movie
    if movie_name not in movie_similarity_df.columns:
        return f"Movie '{movie_name}' not found in the dataset."
    
    similar_movies = movie_similarity_df[movie_name].sort_values(ascending=False)[1:n_recommendations+1]
    
    return similar_movies


```

```python
recommend_movies('Toy Story (1995)')
```
